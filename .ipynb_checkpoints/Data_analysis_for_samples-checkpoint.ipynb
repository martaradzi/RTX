IMPORTS

```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

## Read in a file in json format 

```python
file1 = open('2ksamplex12.txt', 'rb')
data = json.load(file1)
df = pd.DataFrame(data, columns = ['cars', 'overhead'])
file1.close()
df.head()
```

---
# Scale the numebr of cars for plotting reasons

```python
df['scaled10'] = df['cars'].div(10)
df['scaled200'] = df['cars'].div(200)
df['scaled500'] = df['cars'].div(500)
```

# Add a number of features with different window sizes

For this experiment I'll use window size of 500, 750, and 1000 samples as the change of number of cars should be every 250 samples. I'll gather:
- average
- std
- median
- 1st quartile
- 3rd quartile
- variance

I am backfilling the missing values


```python
df['e_average'] = df['overhead'].rolling(window= 500).mean()
df['e_std'] = df['overhead'].rolling(window= 500).std()
df['e_median'] = df['overhead'].rolling(window= 500).median()
df['e_025'] = df['overhead'].rolling(window= 500).quantile(.25)
df['e_075'] = df['overhead'].rolling(window= 500).quantile(.75)
df['e_average'] = df['overhead'].rolling(window= 500).mean()

df['ee_average'] = df['overhead'].rolling(window= 1000).mean()
df['ee_std'] = df['overhead'].rolling(window= 1000).std()
df['ee_median'] = df['overhead'].rolling(window= 1000).median()
df['ee_025'] = df['overhead'].rolling(window= 1000).quantile(.25)
df['ee_075'] = df['overhead'].rolling(window= 1000).quantile(.75)

df['eee_average'] = df['overhead'].rolling(window= 750).mean()
df['eee_std'] = df['overhead'].rolling(window= 750).std()
df['eee_median'] = df['overhead'].rolling(window= 750).median()
df['eee_025'] = df['overhead'].rolling(window= 750).quantile(.25)
df['eee_075'] = df['overhead'].rolling(window= 750).quantile(.75)

df['e_variance'] = df['overhead'].rolling(window= 500).var()
df['ee_variance'] = df['overhead'].rolling(window= 1000).var()
df['eee_variance'] = df['overhead'].rolling(window= 750).var()

df = df.fillna(method ='backfill')
df.head()
```

# Print average overhead of different window sizes against number of cars

```python
df['scaled200'].plot(figsize=(18,10),label ='Number of cars', grid = True).legend()
df['e_average'].plot(figsize=(18,10), label ='Avg window 500', grid = True).legend()
df['ee_average'].plot(figsize=(18,10), label = 'Avg window 1000', grid = True).legend()
df['eee_average'].plot(figsize=(18,10), label = 'Avg window 750', grid = True).legend()
```

```python
df['scaled200'].plot(figsize=(18,10), label = 'Number of cars').legend()
df['e_average'].plot(figsize=(18,10), label ='Avg window 500').legend()
```

```python
df['scaled200'].plot(figsize=(18,10), label = 'Number of cars').legend()
df['ee_average'].plot(figsize=(18,10), label ='Avg window 750').legend()
```

```python
df['scaled10'].plot(figsize=(18,10), label = 'Number of cars').legend()
df['e_variance'].plot(figsize=(18,10), label ='Variance window 500').legend()
df['eee_variance'].plot(figsize=(18,10), label ='Variance window 750').legend()

```
